#### Margin of Error , Upper bound lower bounds with confidence interval 95% and width differencce with 95% confidence interval

```python
p_fm = concat['churn']/(concat['crclscod'])
```

```python
p_fm
```

```python
n=concat['crclscod']
```

```python
# Calculate the standard error

se = np.sqrt(p_fm * (1 - p_fm) / n)
```

```python
se
```

```python
#Now construct the CI using the formulas above. The z-score is 1.96 for a 95% confidence interval.
z_score =1.96
lcb = p_fm - z_score* se #lower bound 
ucb = p_fm + z_score* se #upper bound 
```

```python
lcb
```

```python
ucb
```

```python
#concat['churn_no']=concat['churn']-concat['crclscod']
concat['churn_no']=concat['crclscod']-concat['churn']
concat['churn_rate']=concat['churn']/concat['crclscod']
concat['lower_bound']=lcb
concat['upper_bound']=ucb
concat['width 95% confidence Interval ']=concat['upper_bound']-concat['lower_bound']
concat['margin of error']=se

concat['standard_deviation']=((concat['churn_rate']*(1-concat['churn_rate'])**2+concat['churn_no']*(0-concat['churn_rate'])**2)/concat['churn'])**0.5
concat
concat
```
